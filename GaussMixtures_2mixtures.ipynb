```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as ss
from sklearn.mixture import GaussianMixture
```

```python
df=pd.read_csv('n:\\Probabilidad\\bimodal.csv')
```

```python
#Histogram with Freedman–Diaconis Rule
df.hist(bins='fd',edgecolor='black')
plt.title('Histogram')
plt.xlabel('x')
plt.ylabel('Frequencies')
```

```python
GMM = GaussianMixture(n_components=2, random_state=0).fit(df)
```

```python
GMM.means_
```

```python
GMM.covariances_
```

```python
GMM.weights_
```

```python
x = np.arange(-6., 6., 0.01)
```

```python
pdfs = [p * ss.norm.pdf(x, mu, sd) for mu, sd, p in zip(GMM.means_, GMM.covariances_, GMM.weights_)]
```

```python
density = np.sum(np.array(pdfs), axis=0)
```

```python
density1=density.flatten()
```

```python
plt.plot(x,density1)
```

```python
#Histogram with fd Rule
df.hist(bins='scott',edgecolor='black',density=True)
plt.title('Histogram')
plt.xlabel('x')
plt.ylabel('Frequencies')
plt.plot(x,density1)
```
